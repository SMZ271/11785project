```python
import cmudict
```

```python
dic = cmudict.dict() # Compatible with NLTK
```

```python
def syllable_format(syllables):
  if (len(syllables) == 0):
    return "()"

  output = ""
  for syllable in syllables:
    output += syllable + " "

  output = output[: -1]
  output = "(" + output + ")"

  return output
```

```python
input_path = '/content/gdrive/MyDrive/gpt2/limericks_final2.txt'
output_path = '/content/gdrive/MyDrive/gpt2/limericks_with_syllables.txt'

with open(input_path) as f_in:
  with open(output_path, "w") as f_out:
    lines = f_in.readlines()

    for line in lines:
      if (line == "<|endoftext|>\n"):
        output_line = "<|endoftext|>"

      else:
        words = line.split()
        output_line = ""
        for word in words:
          syllables = dic[word]
          if (len(syllables) == 0):
            syllables = []
          else:
            syllables = syllables[0]
          output_line += word + " " + syllable_format(syllables) + " "
        output_line = output_line[: -1]

      print(output_line)
      f_out.write(output_line + "\n")
```

```python
import pyphen
dic = pyphen.Pyphen(lang='en_US')
```

```python
input_path = '/content/gdrive/MyDrive/gpt2/limericks_final2.txt'
output_path = '/content/gdrive/MyDrive/gpt2/limericks_pythen_syllables.txt'

with open(input_path) as f_in:
  with open(output_path, "w") as f_out:
    lines = f_in.readlines()

    for line in lines:
      if (line == "<|endoftext|>\n"):
        output_line = "<|endoftext|>"

      else:
        words = line.split()
        output_line = ""
        for word in words:
          syllables = dic.inserted(word).split('-')
          for syllable in syllables:
            output_line += syllable + " "
        output_line = output_line[: -1]

      print(output_line)
      f_out.write(output_line + "\n")
```
